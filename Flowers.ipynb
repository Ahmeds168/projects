```python
# This function will plot images in the form of a grid with 1 row and 5 columns where images are placed in each column.
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
import random
from scipy import ndarray

# image processing library
import skimage as sk
from skimage import transform
from skimage import util
from skimage import io
```

```python
#Augmentation
def random_rotation(image_array: ndarray):
    # pick a random degree of rotation between 25% on the left and 25% on the right
    random_degree = random.uniform(-25, 25)
    return sk.transform.rotate(image_array, random_degree)

def call_random_rotation(img_path,j):
    # fetching image
    img_to_augment = sk.io.imread(img_path)
    
    # calling rotation method and saving rotated image
    rotated_img = random_rotation(img_to_augment)
    io.imsave(img_path + "_"+str(j)+"_rotated.jpg", rotated_img)
```

```python
#Augmentation
dirs
for i in classess:
    a=dirs
    totalInstances = len(os.listdir(os.path.join(a,i)))
    #print(1055-totalInstances)
    instances = os.listdir(os.path.join(a,i))
    for j in range(1055-totalInstances):
        imageToAugment = os.path.join(a,i)
        imageToAugment = os.path.join(imageToAugment,instances[j])
        call_random_rotation(imageToAugment,j)
        
    
```

```python
import keras
```

```python
from keras import layers
from keras import models

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(5, activation='softmax'))
```

```python
model.summary()
```

```python
from keras import optimizers

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-4),
              metrics=['acc'])
```

```python
#create validation dir
if(not(os.path.exists("flowers_validation"))):
    os.mkdir("flowers_validation")
    print("dir created")

for i in classess:
    if(not(os.path.exists(os.path.join("flowers_validation",i)))):
        os.mkdir(os.path.join("flowers_validation",i))
        print("dir created")
    
# transfer 200 images from each class
for i in classess:
    tmp = os.path.join(dirs,i)
    print(tmp)
    instances= os.listdir(tmp)
    for j in range(200):
        src = os.path.join(tmp,instances[j])
        dst = os.path.join("flowers_validation", i)
        dst = os.path.join(dst,instances[j])
        shutil.move(src,dst)
        print("moved")
    #for j in range(200):
        
```

```python
from keras.preprocessing.image import ImageDataGenerator

# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        dirs,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use binary_crossentropy loss, we need binary labels
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        "flowers_validation",
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')
```

```python
from PIL import Image
history = model.fit_generator(
      train_generator,
      steps_per_epoch=100,
      epochs=30,
      validation_data=validation_generator,
      validation_steps=50)
```
